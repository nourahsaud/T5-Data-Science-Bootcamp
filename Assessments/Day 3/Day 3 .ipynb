```python
# import 
import pandas as pd
import numpy as np
```

```python
# Read csv
date3 = pd.read_csv('turnstile_160903.csv')
date10 = pd.read_csv('turnstile_160910.csv')
date17 = pd.read_csv('turnstile_160917.csv')

# Concat

data = [date3, date10, date17]

df = pd.concat(data)
df.head()
```

```python
# Change data and time  into a time series

df['DATE'] = pd.to_datetime(df['DATE'])
df['TIME'] = pd.to_datetime(df['TIME'])
df
```

```python
# Max Entries, for each station and date 

unique_station_day = ['C/A', 'UNIT', 'SCP', 'STATION', 'DATE']

df.groupby(unique_station_day)[['ENTRIES']].max().reset_index()
```
